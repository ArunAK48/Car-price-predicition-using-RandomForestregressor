### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###  Load datasets

```python
dataset = pd.read_csv("carPrice.csv")
dataset=dataset.drop(['car_ID'], axis=1)
dataset
```

### convert string column to numeric

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
dataset['fueltype']= label.fit_transform(dataset['fueltype'])
print(dataset)
```

### Splitting x & y

```python
Xdata = dataset.drop('price',axis=1)
numcol=Xdata.select_dtypes(exclude=['object']).columns # new syntax to get all numeric column from datasets
X= Xdata[numcol]
X
```

```python
y =dataset['price']
y
```

### Scaling the independent variables

```python
from sklearn.preprocessing import scale
cols = X.columns
x =pd.DataFrame(scale(X))
print(x)
```

###  Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X, y, test_size=0.20, random_state=0)
```

### Training using randomforest regressor

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

### Evaluate the model

```python
ypred = model.predict(X_test)

from sklearn.metrics import r2_score
r2score = r2_score(y_test, ypred)
print('R2score', r2score*100)
```
